# Bias / Variance

## Import Libraries

```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

## Bias / Variance plots

```python
truth_x = 0
truth_y = 0

size = 20
```

```python
# Gaussian (normal) distribution
# mu + sigma * np.random.standard_normal() = N(mu, sigma)

# b - bias, v - variance
low_b = 0
low_v = 0.1
high_b = 1
high_v = 0.5

rng = np.random.RandomState(42)

x_low_b_low_v = low_b + low_v * rng.standard_normal(size)
y_low_b_low_v = low_b + low_v * rng.standard_normal(size)

x_high_b_low_v = high_b + low_v * rng.standard_normal(size)
y_high_b_low_v = high_b + low_v * rng.standard_normal(size)

x_low_b_high_v = low_b + high_v * rng.standard_normal(size)
y_low_b_high_v = low_b + high_v * rng.standard_normal(size)

x_high_b_high_v = high_b + high_v * rng.standard_normal(size)
y_high_b_high_v = high_b + high_v * rng.standard_normal(size)
```

```python
# four plots

fig, axs = plt.subplots(2, 2, figsize=(8,8))

# top left - low bias low variance
axs[0, 0].scatter(truth_x, truth_y, color="r", marker="X", s=100)
axs[0, 0].scatter(x_low_b_low_v, y_low_b_low_v, marker="o")
axs[0, 0].set_title('Low Bias\nLow Variance', fontsize=14)
axs[0, 0].set_xlim([-2, 2])
axs[0, 0].set_ylim([-2, 2])
axs[0, 0].set_xticklabels([])
axs[0, 0].set_yticklabels([])

# top right - low bias high variance
axs[0, 1].scatter(truth_x, truth_y, color="r", marker="X", s=100)
axs[0, 1].scatter(x_low_b_high_v, y_low_b_high_v, marker="o")
axs[0, 1].set_title('Low Bias\nHigh Variance', fontsize=14)
axs[0, 1].set_xlim([-2, 2])
axs[0, 1].set_ylim([-2, 2])
axs[0, 1].set_xticklabels([])
axs[0, 1].set_yticklabels([])

# bottom left - high bias low variance
axs[1, 0].scatter(truth_x, truth_y, color="r", marker="X", s=100)
axs[1, 0].scatter(x_high_b_low_v, y_high_b_low_v, marker="o")
axs[1, 0].set_title('High Bias\nLow Variance', fontsize=14)
axs[1, 0].set_xlim([-2, 2])
axs[1, 0].set_ylim([-2, 2])
axs[1, 0].set_xticklabels([])
axs[1, 0].set_yticklabels([])

# bottom right - high bias high variance
axs[1, 1].scatter(truth_x, truth_y, color="r", marker="X", s=100)
axs[1, 1].scatter(x_high_b_high_v, y_high_b_high_v, marker="o")
axs[1, 1].set_title('High Bias\nHigh Variance', fontsize=14)
axs[1, 1].set_xlim([-2, 2])
axs[1, 1].set_ylim([-2, 2])
axs[1, 1].set_xticklabels([])
axs[1, 1].set_yticklabels([])
```

```python
# one plot for four situations

plt.figure(figsize=(8, 8))
plt.scatter(truth_x, truth_y, 
            color="m", marker="X", s=1000, label="True Value")

plt.scatter(x_low_b_low_v, y_low_b_low_v,
            color="g", marker="o", s=50, label="Low Bias Low Variance")
plt.scatter(x_low_b_high_v, y_low_b_high_v,
            color="y", marker="o", s=50, label="Low Bias High Variance")
plt.scatter(x_high_b_low_v, y_high_b_low_v,
            color="b", marker="o", s=50, label="High Bias Low Variance")
plt.scatter(x_high_b_high_v, y_high_b_high_v,
            color="r", marker="o", s=50, label="High Bias High Variance")

plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.legend(loc='best', fontsize=14)

plt.xlim(-2, 2)
plt.ylim(-2, 2)
```
